# プログラミング（Python）で統計をやろう！

```python
# データを扱うライブラリーをインポート
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# データを作る
data = {
    'study':[2,8,0,2,0,0,4,0,0,1,1,0,3,1,0,1,2,2,5,1,0,3,30,3,1],
    'reitaku':[1,10,7,5,5,5,6,10,5,8,7,6,9,4,9,8,5,7,9,5,5,8,9,8,7]
}
```

```python
# データをテーブル化し、変数に入れる
df = pd.DataFrame(data)
```

```python
# テーブルを表示
df
```

## 散布図

```python
# 散布図を作る
plt.scatter(df['study'], df['reitaku'])

# Set plot labels and title
plt.xlabel('Study')
plt.ylabel('Reitaku')
plt.title('Scatter Plot of Study vs Reitaku');
```

## 統計を出そう

平均値

```python
# studyの平均値
df['study'].mean()
```

```python
# reitakuの平均値は？
```

標準偏差

```python
# studyの標準偏差
df['study'].std(ddof=0)
```

```python
# reitakuの標準偏差は？
```

共分散

```python
# studyとreitakuの共分散
df['study'].cov(df['reitaku'],ddof=0)
```

相関係数

```python
# studyとreitakuの相関係数
df['study'].corr(df['reitaku'])
```
